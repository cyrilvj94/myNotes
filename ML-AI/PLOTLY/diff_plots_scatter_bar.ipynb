__Plotly update_traces and update_layout()__

```python
fig = make_subplots(1,2)
# subplot 1
trace1 = go.Scatter(x = normal(size = (100,)), y = uniform(size = (100,)), name = 'plot 1 scatter', mode = 'markers',
                   marker = dict(symbol = 0, opacity = 0.7, size = uniform(0,4 , size = (100,)), sizeref = 0.2
                                                        ,color = 'cyan', ))
fig.add_trace(trace = trace1, row = 1, col = 1)
fig.add_trace(go.Bar(x = np.arange(0, 2, 0.25), y = np.arange(0, 2, 0.25),
                             marker_color = 'green', marker_opacity = 0.4, name = 'plot 1 Bar'), 1, 1)

#subplot 2
trace2 = go.Scatter(x = uniform(10, 30, size = 100), y = uniform(2, 20 , size = 100), name = 'Scatter plot ', 
                   mode = 'markers', marker = dict(symbol = 1, opacity = 0.7, size = uniform(0, 4,size = (100,) )
                                                   ,sizeref = 0.2, color = 'red' ))
fig.add_trace(trace = trace2, row  = 1, col = 2)
fig.add_trace(go.Bar(x = np.arange(10, 20, 0.25), y = np.arange(10, 20, 0.25),
                             marker_color = 'orange', marker_opacity = 0.4, name = 'plot 2 bar'), 1,  2)

fig.show()
```

```python
fig.update_layout(title_text = 'updating plots', legend_bgcolor = 'rgba(2,4,5,0.3)', 
                  title_font_family = 'Arial', title_font_size = 19, title_font_color = 'Black',
                  xaxis_title_text = 'Xaxis of plot 1', xaxis_title_font_size = 20,
                 yaxis_title_text = 'Y xis of plot 1', yaxis_title_font_size = 20,)
fig.show()
```

__customising plots__

```python
iris = px.data.iris()
iris.head()
```

```python
fig = px.scatter(iris, x = 'sepal_length', y = 'sepal_width', color = 'species',hover_data = ['petal_length'], 
           size = 'petal_width',
        labels = {'sepal_width': "Sepal Widht in cms", 'sepal_length': 'Sepal Length in cms'},)
fig.update_traces( marker_opacity = 0.7, )
fig.update_layout(title = dict(text = 'Variation of iris flower', font_size = 20), 
                  title_x = 0.5,title_y = 0.9, title_font_color = 'red',
                  legend_title_text = 'Flower Species',legend_bgcolor = 'rgba(0,255,255,0.1)',legend_bordercolor = 'black', 
                  legend_borderwidth  = 3, legend_x = 0.01, legend_tracegroupgap = 0.01)

#if you want to change the name of the legend items

for i , j in zip(['Setosa: Good for medicine' , 'Vesicolor: Good for cancer treatment', 'virginica: Homeopathic medicine']
                 ,['setosa', 'versicolor', 'virginica']):
    fig.update_traces(name = i, selector = {'name':j})
    
#changing the size of the figure , NOTE HEIGHT AND WIDTH ARE IN PIXELS

fig.update_layout(height = 600, width =1200)

fig.show()
```

```python
fig = px.scatter(iris, x = 'sepal_length', y = 'sepal_width', color = 'species',hover_data = ['petal_length'], 
           size = 'petal_width')
fig.update_layout(title_text = 'Flower species variation', title_font_size = 15, title_font_color = 'red')

for i , j in zip(['Setosa: Good for medicine' , 'Vesicolor: Good for cancer treatment', 'virginica: Homeopathic medicine']
                 ,['species=setosa', 'species=versicolor', 'species=virginica']):
    fig.update_traces(name = i, selector = {'legendgroup':j})
    
fig.update_layout(legend_bgcolor = 'aqua', legend_font_color = 'red', legend_font_size = 15, legend_title_text = 'Species of flower')
    
fig.show()
```

```python
data = px.data.gapminder()
```

```python
fig = make_subplots(rows = 3, cols = 1, shared_xaxes= True)
#trace1 = px.line(data[data.country == 'India'], x = 'year', y = 'pop')

# Trace 1
trace_1 = go.Scatter(x = data[data.country == 'India']['year'], y = data[data.country == 'India']['pop'], mode = 'lines',
                    name = 'Population')
fig.add_trace(trace = trace_1, row = 1,col =  1 )


# Trace 2
trace_2 = go.Scatter(x = data[data.country == 'India']['year'], y = data[data.country == 'India']['gdpPercap'], mode = 'lines',
                    name = 'gdp')
fig.add_trace(trace = trace_2, row = 2,col =  1 )


# Trace 3
trace_3 = go.Scatter(x = data[data.country == 'India']['year'], y = data[data.country == 'India']['lifeExp'], mode = 'lines',
                    name = 'life exp')
fig.add_trace(trace = trace_3, row = 3,col =  1 )

# Updating the layout

fig.update_layout(height = 1000, title_text = 'Variation of pop gdp and life exp over various year in india',
                  title_font_size = 20, title_font_color = 'black', legend = dict(title_text = 'graph variables'
                                                                                  ,bgcolor = 'rgb(0,255,255,0.3)', 
                                                                                 borderwidth = 3, x = 1.1, y = 1.1))
fig.update_xaxes(title_text = 'Year', title_font_color = 'mediumvioletred', title_font_size = 10)


#Setting axis names
fig.update_yaxes(title_text = 'Population in billions', row = 1, col = 1,  title_font_size = 12 )
fig.update_yaxes(title_text = 'gdp', row = 2, col = 1 )
fig.update_yaxes(title_text = 'life_exp in years', row = 3, col = 1 , title_font_size = 12)

# Changing colours
[
fig.update_traces(marker_color = 'rgba({},{},{},{})'.format(uniform(0,255,1)[0],uniform(0,255,1)[0],uniform(0,255,1)[0]
                                                             ,uniform(0,255,1)[0]), selector = {'xaxis':i}) 
    for i in ['x', 'x2', 'x3']

]

fig.show()
```

__When more than one row share the same x value they are stacked on top of one another by default
 barmode = 'group' will be properly applied only for categorical legends__

```python
#fig = make_subplots(rows = 2, cols = 1)
"""
Legend is not categorical
"""
data.year = data.year.astype(np.int32)
fig = px.bar(data.query('country == "India" or country =="China"'), x = 'country', y = 'pop', color = 'year',
             hover_data = ['year'], barmode = 'group', height = 400)
fig.update_layout(barmode = 'group')
               
fig.show()
```

```python
#fig = make_subplots(rows = 2, cols = 1)
"""
Legend is categorical
"""
data.year = data.year.astype(np.str)
fig = px.bar(data.query('country == "India" or country =="China"'), x = 'country', y = 'pop', color = 'year',
             hover_data = ['year'], barmode = 'group', height = 400)
fig.update_layout(barmode = 'group')
               
fig.show()
```
