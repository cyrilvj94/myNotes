# EXCERCISE 7 SUBPLOTS

In the all the excercises done so far all plots are created one over the other , it is sometimes often best to create these overlapped plots seperately, this is where plt.subplots() comes into use

```python
plt.scatter( np.arange(250) , d.iloc[:250,0], s = 100, marker = 'o' 
            , edgecolor = 'k', color = 'g', alpha = 0.5, linewidths = 0.25)
plt.scatter( np.arange(250) , d.iloc[:250,1], s = 100, marker = 'o' 
            , edgecolor = 'k', color = 'r', alpha = 0.5, linewidths = 0.25)
plt.scatter( np.arange(250) , d.iloc[:250,2], s = 100, marker = 'o' 
            , edgecolor = 'k', color = 'b', alpha = 0.5, linewidths = 0.25)
```

```python



fig,(ax1,ax2,ax3) = plt.subplots(nrows = 3,ncols = 1, sharex = True, sharey  = True, figsize = (10,10))
ax1.scatter( np.arange(250) , d.iloc[:250,0], s = 500, marker = 'o' 
            , edgecolor = 'k', color = 'g', alpha = 0.5, linewidths = 0.25,label = 'x')


ax2.scatter( np.arange(250) , d.iloc[:250,1], s = 500, marker = 'o' 
            , edgecolor = 'k', color = 'r', alpha = 0.5, linewidths = 0.25, label = 'y' )

ax3.scatter( np.arange(250) , d.iloc[:250,2], s = 500, marker = 'o' 
            , edgecolor = 'k', color = 'b', alpha = 0.5, linewidths = 0.25, label = 'z')

ax3.legend(loc = 3)

ax1.legend(loc = 3)
ax2.legend(loc = 3)

ax1.set_xlabel('weight')
ax2.set_xlabel('weight')
ax3.set_xlabel('weight')

ax3.set_ylabel('age')

plt.tight_layout()
plt.show()
```

PLOTTING AUTOCORRELATION AND PARTIAL AUTO CORRELATION

```python
ts = pd.read_csv(r'C:\Users\DELL\Desktop\ai\ineuron.ai\timeSeries\daily-total-female-births-CA.csv')
ts[ts.columns[0]] = pd.to_datetime(ts[ts.columns[0]])
ts.set_index('date', inplace = True)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
fig1, ax1 = plt.subplots(ncols = 1, nrows =2 , figsize = (10,10))

fig1 = plot_acf(ts, ax = ax1[0], color = 'r')
ax1[0].set_xlabel('no oflags')
ax1[0].set_ylabel('atocorrelation values')

fig1 = plot_pacf(ts, ax = ax1[1])
ax1[1].set_xlabel('no of lags')
ax1[1].set_ylabel('partial atocorrelation values')
```
