```python
# Looping the entire taskd
from sklearn.datasets import load_iris
data = load_iris()

plt.style.use('fivethirtyeight')

#Load iris dataset
data = pd.DataFrame(data = np.concatenate([data.data, data.target.reshape(-1 ,1)], axis = 1),
             columns = data.feature_names + ['Target'])
data = data.rename(columns = dict(zip(data.columns, ['Sepal_length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])))

def create_plot(ax,data, x, y, target = 'Target', cmap = 'flare' ):
    """
    Creates a line plot in given axes , ax 
    """
    temp = ax.scatter(data[x] ,data[y], c = data[target], cmap = cmap)
    ax.set_title(f'{y} Vs {x}'.replace('_', ' '))
    ax.set_xlabel(x.replace('_', ' '))
    ax.set_ylabel(y.replace('_', ' '))
    cbar = plt.colorbar(temp, ax = ax)
    cbar.set_ticks([0, 1, 2])
    cbar.set_ticklabels(['setosa', 'versicolor', 'virginica'])
 

fig , axes  = plt.subplots(2,2, figsize = (15,12), dpi = 300)

for ax, x, y in list(zip(axes.flatten(), ['Sepal_length', 'Sepal_Width', 'Petal_Length', 'Sepal_Width'], \
                       ['Petal_Length', 'Petal_Width', 'Petal_Width', 'Sepal_length'])):
    create_plot(ax, data, x, y, cmap = plt.colormaps()[88])

fig.suptitle('IRIS FLOWER PLOTS', y = 0.92)
plt.show()
fig.savefig('iris flower plots.jpg')
```

```python
# Looping the entire taskd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
data = load_iris()

plt.style.use('fivethirtyeight')

#Load iris dataset
data = pd.DataFrame(data = np.concatenate([data.data, data.target.reshape(-1 ,1)], axis = 1),
             columns = data.feature_names + ['Target'])
data = data.rename(columns = dict(zip(data.columns, ['Sepal_length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])))
data.Target = data.Target.replace({0:'Setosa', 1:'Versicolor', 2:'Virginica'})

def create_plot(ax,data, x, y, target = 'Target', cmap = 'cool' ):
    """
    Creates a line plot in given axes , ax 
    """
    temp = sns.scatterplot(x = x, y = y, data = data, hue = target, palette= cmap, ax = ax,  s = 200, marker = '8',
                           alpha = 0.7, linewidths =3)
    ax.legend(loc = 'best', title = 'Species', frameon =  True, facecolor = 'y' , framealpha = 0.3,fontsize = 15)
    ax.set_title(f'{y} Vs {x}'.replace('_', ' '))
    ax.set_xlabel(x.replace('_', ' '))
    ax.set_ylabel(y.replace('_', ' '))
    
 

fig , axes  = plt.subplots(2,2, figsize = (15,12), dpi = 300)

for ax, x, y in list(zip(axes.flatten(), ['Sepal_length', 'Sepal_Width', 'Petal_Length', 'Sepal_Width'], \
                       ['Petal_Length', 'Petal_Width', 'Petal_Width', 'Sepal_length'])):
    create_plot(ax, data, x, y, )

fig.suptitle('IRIS FLOWER PLOTS', y = 0.92, fontsize = 25, color = 'green')
plt.show()
fig.savefig('iris flower plots seaborn.jpg')
```

## Pie Chart

```python
fifa_data = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')
```

```python
plt.style.use('classic')
fig, ax =  plt.subplots(1,2 , figsize = (12,6))

#Foot used 

ax[0].pie(fifa_data['Preferred Foot'].value_counts().values, autopct = '%.2f' , wedgeprops = {'linewidth': 3, 'hatch':'o'}, 
        textprops = {'fontsize':20, 'color':'w'},
        colors = [(1,0.5,0.5,0.5), (0.3,0.3,0.3,0.3)], shadow = True, 
        labels = fifa_data['Preferred Foot'].value_counts().index)
ax[0].legend(loc = 'best', fontsize = 15,)
ax[0].set_title('Foot used by FIFA players')

#Data for Nationality of players

nation_data = fifa_data.Nationality.value_counts()
others = nation_data.count() - nation_data[nation_data<100].count()
nation_data = nation_data[nation_data>100]
nation_data = nation_data.append(pd.Series(data = [others], index = ['Others']))

#Pie chart for Nationality
ax[1].pie(x = nation_data.values, labels = nation_data.index)
plt.show()
```
