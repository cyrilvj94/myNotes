```python
# Create a connection object 
con = sqlite3.connect('chinook.db')
# Create a cursor to execute statements;
cursor = con.cursor()
```

__Sql statements can be executed using cursor.execute() command__

```python
print(dir(cursor))
```

```python
#show tables
for i in cursor.execute('select name from sqlite_master where type = "table"'):
    print(i)
```

```python
for i in cursor.execute("select * from albums"):
    print(i)
```

```python
cursor.execute('create table  if not exists newTable (id integer, name text)')
#show tables
for i in cursor.execute('select name from sqlite_master where type = "table"'):
    print(i)
con.commit()
```

```python
cursor.execute("Insert into newTable (id , name)  values (2, 'cyril'), (1, 'karthik'), (3, 'roshan'), ( 4, 'nkunku') ")
con.commit()
```

```python
for i in cursor.execute("select * from newTable"):
    print(i)
```

```python
con.close()
```

```python
import numpy as np
data = np.random.randint(10, 100, (100,5)).tolist()
conn = sqlite3.connect('test.db')
cursor = conn.cursor()
try:
    cursor.execute('create table  newTable (col1 integer, col2 integer, col3 integer, col4 integer , col5 integer) ')
except :
    cursor.execute('drop table newTable')
    cursor.execute('create table  newTable (col1 integer, col2 integer, col3 integer, col4 integer , col5 integer) ')
stmt = 'insert into newTable values (?, ?, ? , ?, ?)'
cursor.executemany(stmt, data)
cursor.execute('select * from newTable')
cursor.fetchall()
```
